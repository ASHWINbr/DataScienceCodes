```python
X = df_train_cleaned.loc[:,df_train_cleaned.columns!='label']
Y = df_train_cleaned[['label']]

print(X.shape,Y.shape)
```

```python
test= df_test_cleaned.loc[:,:]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=104,train_size=0.9,shuffle=True)
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
std_scaler = StandardScaler()
```

```python
X_train.values[:] = std_scaler.fit_transform(X_train)
X_test.values[:] = std_scaler.fit_transform(X_test)
```

```python
model = GaussianNB()
# train the model
model.fit(X_train, Y_train)
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(test)
```

```python
model.score(X_test,Y_test)
```
